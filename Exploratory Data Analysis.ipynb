### Extracting X Resolution and Y Resoltuion

```python
# Splitting the the x and y resoltion on x

resolution=df['ScreenResolution'].str.split('x',n=1,expand=True)
resolution.head()
```

```python
df['XRes']=resolution[0]
df['YRes']=resolution[1]
df.head()
```

```python
''' We can observe that XRes column has some extra details along with x resolution.
    We need to extract the x resolution from XRes , we will use regex to extract this'''

'''so firstly we will replace all the "," with "" and then find all numbers
from that string as "\d+\.?\d+",\d means that integer number and \.? 
all the numbers which come after an number and \d+ the string must end with number'''

df['XRes']=df['XRes'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])
```

```python
df.head()
```

```python
# converting XRes and YRes columns into int type

df['XRes']=df['XRes'].astype('int')
df['YRes']=df['YRes'].astype('int')
df.info()
```

```python
# Heatmap for looking into correlation

plt.figure(figsize=(15,7))
sn.heatmap(df.corr(),annot=True,cmap='plasma')
```

```python
df.corr()['Price']
```

**From the correlation plot we observed that as the X_res and Y_res is increasing,the price of the laptop is also increasing,so X_res and Y_res are positively correlated and they are giving much information,so that is the reason why i had splitted Resolution column into X_res and Y_res columns respectively**

**So to make things good,we can create a new column named `PPI{pixels per inch}`,now  as we saw from the correlation plot that the `X_res and Y_res` are having much collinearity,so why not combine them with `Inches` which is having less collinearity,so we will combine them as follows ,so here is the formula of how to calculate `PPI` {pixels per inch}**

$$
    PPI(pixels per inch) = \frac{\sqrt{X Resolution^2+Y Resolution^2}}{inches}
$$

```python
df['PPI']= ((df['XRes']**2 + df['YRes']**2)**0.5/df['Inches']).astype('float')
df.head()
```

```python
df.corr()['Price']
```

**So as we observe from the correlation data that the PPI is having good correlation,so we will be using that,as that is a combination of 3 features and that gives collective results of 3 columns,so we will drop Inches,X_res,Y_res as well**

```python
# Dropping ScreenResolution Inches XRes and YRes

df.drop(columns=['ScreenResolution','Inches','XRes','YRes'],inplace=True)
df.head()
```

### Memory Column Analysis

**Memory columns holds important ans multiple data we need to extract the data and segregate it. For example 128GB SSD +  1TB HDD**

```python
# values count and variation of data in Memory

df['Memory'].value_counts()
```

**There are four types of memory i.e. SSD, HDD, Flash Storage and hybrid or combination of them we will extract for ech of them and create a new column for each type of memmory**

```python
# Removing decimal(.) example 1.0 TB will become 1 TB

df['Memory']=df['Memory'].astype('str').replace('\.0','',regex=True)

# Replacing GB with blank

df['Memory']=df['Memory'].str.replace('GB','')

# Replacing TB with "000" 

df['Memory']=df['Memory'].str.replace('TB','000')
```

```python
df["Memory"].value_counts()
```

```python
#splitting the column data on '+' and storing in dummy variable

dummy=df['Memory'].str.split('+',n=1,expand=True)
dummy
```

```python
# we will strip all white spaces and inclue first column of dummy variable into the data frame

df['first']=dummy[0]
df['first']=df['first'].str.strip()
df.sample(10)
```

```python
#Creating Layer SSD,HDD,Flash & Hybrid Column for our calculation

def applychanges(value):
    
    df['Layer1'+value]=df['first'].apply(lambda x:1 if value in x else 0)
    
    
listtoapply = ['HDD','SSD','Hybrid','FlashStorage'] 

for value in listtoapply:
    applychanges(value)
    
    
df.head()   
```

```python
# Removing all character from first column keeping only numbers

df['first']=df['first'].str.replace(r'\D','')
df.head()
```

```python
# Simillarly we will do for dummy second column

df['Second']=dummy[1]
df['Second']=df['Second'].str.strip()

df.head()
```

```python
#Creating 2nd Layer for SSD,HDD,Flash & Hybrid Column for our calculation

def valuechange(value):
    
    df['Layer2'+value]=df['Second'].apply(lambda x:1 if value in x else 0)
    
    
listtoapply= ['SSD','HDD','FlashStorage','Hybrid']
#Replacing None value with 0

df['Second']=df['Second'].fillna('0')

for value in listtoapply:
    valuechange(value)
    
df.head()
```

```python
# Removing all character from Second column, keeping only numerical

df['Second']=df['Second'].str.replace(r'\D','')
df.head()
```

```python
# convertinf fisrt and second column as integer

df['first']=df['first'].astype('int64')
df['Second']=df['Second'].astype('int64')

df.info()
```

```python
# Now storing the memory value in respective column, creating columns and stroing on them

df["HDD"]=(df["first"]*df["Layer1HDD"]+df["Second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["Second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["Second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1FlashStorage"]+df["Second"]*df["Layer2FlashStorage"])

df.head()
```

```python
# Removing the unwanted column 'first', 'Second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
#       'Layer1FlashStorage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
#       'Layer2FlashStorage'

df.drop(columns=['first', 'Second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1FlashStorage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2FlashStorage'],inplace=True)
```

### Operating system analysis (OpSys)

```python
df['OpSys'].value_counts()
```

```python
sn.barplot(df['OpSys'],df['Price'])
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
df['OpSys'].unique()
```

**We will club all type of windows os as windows and all types mac os as mac and linux and other types of os as others** 

```python
# club {Windows 10,Windows 7,Windows 7 S}-->Windows
# club {macOS,mac OS X}--> mac
#linux --> Linux
# else return Others


def setcategory(text):
    
    if text=='Windows 7' or text=='Windows 10' or text=='Windows 10 S':
        return 'Windows'
    
    elif text=='macOS' or text=='Mac OS X':
        return 'Mac'
    
    elif text=='Linux':
        return 'Linux'
    
    else:
        return 'Others'
    

df['OpSys']= df['OpSys'].apply(lambda x: setcategory(x))
df.head()
```

```python
# Barplot of Opsys wrt price

sn.barplot(df['OpSys'],df['Price'])
plt.show()
```

**Mac OS system are expensive and also increases price of laptop, windows are common operating system and others and linux operating system laptop proce are low and affordable**
